```python
#convertendo data - str > datetime
df_pedidos['data'] = pd.to_datetime(df_pedidos['data'], format='%d/%m/%Y %H:%M:%S')
```

```python
#obter apenas data, sem time
df_pedidos['data_aux'] = pd.to_datetime(df_pedidos['data']).dt.date
```

```python
# Agrupar por ID do pedido e obter a primeira ocorrência de cada data para cada pedido
df_pedidos_unicos = df_pedidos.groupby('id').first().reset_index()

# Contar pedidos por data
contagem_por_data = df_pedidos_unicos.groupby('data_aux').size().reset_index(name='contagem')

# Ordenar pela contagem em ordem decrescente
print(contagem_por_data.sort_values('contagem', ascending=False))
```

```python
#criando nova coluna - dia da semana
df_pedidos['dia_semana'] = df_pedidos['data'].dt.day_name()

df_pedidos.head()
```

```python
# Agrupar por ID do pedido e obter a primeira ocorrência de cada data para cada pedido
df_pedidos_unicos = df_pedidos.groupby('id').first().reset_index()

# Agrupar por 'dia_semana' e contar o número de pedidos únicos
contagem_por_dia = df_pedidos_unicos.groupby('dia_semana').size().reset_index(name='contagem')

# Ordenar a contagem em ordem decrescente
print(contagem_por_dia.sort_values('contagem', ascending=False))
```

```python
# Frequência esperada se os pedidos fossem uniformemente distribuídos
frequencia_esperada = len(df_pedidos_unicos) / 7

print(frequencia_esperada)
```

```python
# Extrair a lista de contagens observadas e a lista de dias da semana
contagens_observadas = contagem_por_dia['contagem']
dias_da_semana = contagem_por_dia['dia_semana']

# Calcular a frequência esperada (assumindo uma distribuição uniforme)
frequencia_esperada = contagens_observadas.sum() / len(dias_da_semana)
frequencias_esperadas = [frequencia_esperada] * len(dias_da_semana)

# Realizar o teste qui-quadrado
chi2, p_valor = stats.chisquare(contagens_observadas, f_exp=frequencias_esperadas)

# Exibir os resultados
print(f"Estatística qui-quadrado: {chi2}")
print(f"Valor-p: {p_valor}")
```

```python
if p_valor < 0.05:
    print("Rejeitamos a hipótese nula: há evidência de sazonalidade nos dias da semana.")
else:
    print("Não rejeitamos a hipótese nula: não há evidência de sazonalidade nos dias da semana.")
```
